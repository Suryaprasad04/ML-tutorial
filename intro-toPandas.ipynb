```python
car_sales_missing=pd.read_csv("car-sales-missing-data.csv")
car_sales_missing
```

```python
car_sales_missing["Odometer"].mean()
```

```python
car_sales_missing["Odometer"]=car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean())
```

```python
car_sales_missing
```

```python
car_sales_missing["Odometer"].mean()
```

```python
car_sales_missing_dropped=car_sales_missing.dropna()
```

```python
car_sales_missing_dropped
```

```python
#columns from series
seats_column=pd.Series([5,5,5,5,5])

#New columnn called seats
car_sales["Seats"]=seats_column
car_sales
```

```python
car_sales["Seats"]=car_sales["Seats"].fillna(5)
car_sales
```

```python
#column from python list
fuel_economy=[6.7,8.8,5.6,7.8,5.7,8.9,9.5,5.9,8.9,6.7]#should be of same length if we derive from list
car_sales["Fuel per 100km"]=fuel_economy
car_sales
```

```python
car_sales["Total fuel used (L)"]=car_sales["Odometer (KM)"]/100*car_sales["Fuel per 100km"]
car_sales
```

```python
#create a column from  a single value
car_sales["Number of wheels"]=4
```

```python
car_sales
```

```python
car_sales=car_sales.drop("Total fuel used (L)",axis=1)
car_sales
```

```python
car_sales["Passed road safety"]=True
car_sales
```

```python
car_sales.sample(frac=0.5)#50 per of data and not in order
```

```python
car_sales_shuffeled=car_sales.sample(frac=1)#100 per of data and not in order
car_sales_shuffeled
```

```python
#only select 20% of data
car_sales_shuffeled.sample(frac=0.2)
```

```python
car_sales_shuffeled.reset_index()#cretes another index with the old one
```

```python
car_sales_shuffeled.reset_index(drop=True,inplace=True)#resets inplace
car_sales_shuffeled
```

```python
car_sales["Odometer (KM)"]=car_sales["Odometer (KM)"].apply(lambda x:x/1.6)
car_sales
```
